```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
#GAN Implementation
# Hyperparameters
latent_dim = 128
hidden_dim = 256
batch_size = 128
epochs = 200
#separate learning rates for D and G

# Loading the MNIST dataset
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize([0.5], [0.5])]) #normalising it
dataset = torchvision.datasets.MNIST(root="./data", train=True, transform=transform, download=True)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Generator
class Generator(nn.Module):
    def __init__(self, latent_dim, hidden_dim):
        super(Generator, self).__init__()
        self.model = nn.Sequential(nn.Linear(latent_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 28 * 28),
            nn.Tanh()) 
    def forward(self, z):
        img = self.model(z)
        img = img.view(img.size(0), 1, 28, 28)
        return img

# Discriminator
class Discriminator(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(nn.Linear(input_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, 1),
            nn.Sigmoid())
    def forward(self, img):
        img_flat = img.view(img.size(0), -1)
        validity = self.model(img_flat)
        return validity

# Initializing generator and discriminator
generator_GAN = Generator(latent_dim, hidden_dim).to(device)
discriminator_GAN = Discriminator(28 * 28, hidden_dim).to(device)
criterion = nn.BCELoss() #Binary Cross Entropy Loss
optimizer_G = optim.Adam(generator_GAN.parameters(), lr=0.0001, betas=(0.5, 0.999))  #ADAM Optimizer fro Generator
optimizer_D = optim.Adam(discriminator_GAN.parameters(), lr=0.0002, betas=(0.5, 0.999))  #ADAM Optimizer for Discriminator
G_losses = [] #collecting the losses
D_losses = []
fixed_z = torch.randn(16, latent_dim).to(device) #latent variable

for epoch in range(epochs):
    for i, (imgs, _) in enumerate(dataloader):
        imgs = imgs.to(device)
        imgs += 0.05 * torch.randn_like(imgs).to(device)  #adding noise to real image fro robustness
        batch_size = imgs.size(0)

        z = torch.randn(batch_size, latent_dim).to(device) # Training Discriminator
        real_labels = torch.ones(batch_size, 1).to(device)
        fake_labels = torch.zeros(batch_size, 1).to(device)
        outputs = discriminator_GAN(imgs) # Real images
        d_loss_real = criterion(outputs, real_labels)
        fake_imgs = generator_GAN(z) # Fake images
        outputs = discriminator_GAN(fake_imgs.detach())
        d_loss_fake = criterion(outputs, fake_labels)
        d_loss = d_loss_real + d_loss_fake # Discriminator loss
        optimizer_D.zero_grad()
        d_loss.backward()
        optimizer_D.step()

        z = torch.randn(batch_size, latent_dim).to(device) # Training Generator
        fake_imgs = generator_GAN(z)
        outputs = discriminator_GAN(fake_imgs)
        g_loss = criterion(outputs, real_labels)
        optimizer_G.zero_grad()
        g_loss.backward()
        optimizer_G.step()

        G_losses.append(g_loss.item())
        D_losses.append(d_loss.item())

    print(f"Epoch [{epoch+1}/{epochs}]  Loss D: {d_loss.item():.4f}, Loss G: {g_loss.item():.4f}")
    if (epoch + 1) % 20 == 0:  # Visualizing generator output fro every 20 epochs
        with torch.no_grad():
            generated_imgs = generator_GAN(fixed_z).cpu()
            generated_imgs = (generated_imgs + 1) / 2  # Rescale to [0, 1]
            grid = torchvision.utils.make_grid(generated_imgs, nrow=4, padding=2, normalize=True)
            plt.figure(figsize=(8, 8))
            plt.imshow(np.transpose(grid, (1, 2, 0)))
            plt.title(f"Generated Images at Epoch {epoch+1}")
            plt.axis("off")
            plt.show()
```

```python
#plotting loss curves
plt.figure(figsize=(10, 5))
plt.plot(G_losses, label="Generator Loss")
plt.plot(D_losses, label="Discriminator Loss")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
#Latent Space Interpolation in GANs

generator_GAN.eval()
with torch.no_grad():
    z1 = torch.randn(1, latent_dim).to(device) #generating a random image to interpolate from
    z2 = torch.randn(1, latent_dim).to(device) #generating a random image to interpolate to
    num_steps = 12 #number of interpolation steps to visualise
    interpolation_imgs = []

    img1 = generator_GAN(z1).cpu()
    img1 = (img1 + 1) / 2  
    img2 = generator_GAN(z2).cpu()
    img2 = (img2 + 1) / 2  

    plt.figure(figsize=(4, 4)) #visualising the generated random images
    plt.subplot(1, 2, 1)
    plt.imshow(img1.squeeze(0).squeeze(0), cmap='gray')
    plt.title("Generated Image 1")
    plt.axis("off")

    plt.subplot(1, 2, 2) 
    plt.imshow(img2.squeeze(0).squeeze(0), cmap='gray')
    plt.title("Generated Image 2")
    plt.axis("off")
    plt.show()

    for alpha in np.linspace(0, 1, num_steps):
        z_interp = (1 - alpha) * z1 + alpha * z2 #linear inteerpolation 
        generated_img = generator_GAN(z_interp).cpu()
        generated_img = (generated_img + 1) / 2  
        interpolation_imgs.append(generated_img.squeeze(0))

    grid_interp = torchvision.utils.make_grid(interpolation_imgs, nrow=num_steps, padding=2, normalize=True) #visualising the interpolation
    plt.figure(figsize=(20, 5))
    plt.imshow(np.transpose(grid_interp, (1, 2, 0)))
    plt.title("Latent Space Interpolation between Two Generated Images")
    plt.axis("off")
    plt.show()
```

```python
# DCGAN for Image Generation
#hyperparameters
latent_dim = 1024
batch_size = 128
#have kept separate learning rates for D and G
epochs = 100

# Loading the CIFAR-10 dataset
transform = transforms.Compose([transforms.Resize(32), transforms.ToTensor(), transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])])
dataset = torchvision.datasets.CIFAR10(root="./data", train=True, transform=transform, download=True)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

gen_layers = nn.Sequential(nn.ConvTranspose2d(latent_dim, 512, kernel_size=2, stride=1, padding=0, bias=False), #Generator architecture
    nn.BatchNorm2d(512),
    nn.ReLU(True),
    nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.ReLU(True),
    nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.ReLU(True),
    nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(64),
    nn.ReLU(True),
    nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1, bias=False),
    nn.Tanh())
def generator_forward(z):
    return gen_layers(z)

disc_layers = nn.Sequential(nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1, bias=False),  #discriminator architecture
    nn.LeakyReLU(0.2, inplace=True),
    nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.LeakyReLU(0.2, inplace=True),
    nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.LeakyReLU(0.2, inplace=True),
    nn.Conv2d(256, 512, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(512),
    nn.LeakyReLU(0.2, inplace=True),
    nn.Conv2d(512, 1, kernel_size=2, stride=1, padding=0, bias=False),
    nn.Sigmoid())
def discriminator_forward(img):
    validity = disc_layers(img)
    return validity.view(-1, 1)

generator_DCGAN = gen_layers.to(device)  # Initializing generator and discriminator
discriminator_DCGAN = disc_layers.to(device)

criterion = nn.BCELoss()  #Binary Cross Entropy Loss
optimizer_G = optim.Adam(generator_DCGAN.parameters(), lr=0.0002, betas=(0.5, 0.999))   #ADAM optimizer for G, have mentioned LR here
optimizer_D = optim.Adam(discriminator_DCGAN.parameters(), lr=0.00002, betas=(0.5, 0.999))   #ADAM optimizer for D, have mentioned LR here

G_losses = []
D_losses = []
fixed_z = torch.randn(16, latent_dim, 1, 1).to(device)
smooth_label_value = 0.9  # Value for label smoothing for better training

for epoch in range(epochs):
    for i, (imgs, _) in enumerate(dataloader):
        imgs = imgs.to(device)
        imgs += 0.005 * torch.randn_like(imgs).to(device)
        batch_size = imgs.size(0)
        
        z = torch.randn(batch_size, latent_dim, 1, 1).to(device) # Training the discriminator
        real_labels = torch.full((batch_size, 1), smooth_label_value, device=device)  # Applying label smoothing 
        fake_labels = torch.zeros(batch_size, 1).to(device)
        outputs = discriminator_forward(imgs)
        d_loss_real = criterion(outputs, real_labels)
        fake_imgs = generator_forward(z)
        outputs = discriminator_forward(fake_imgs.detach())
        d_loss_fake = criterion(outputs, fake_labels)
        d_loss = d_loss_real + d_loss_fake
        optimizer_D.zero_grad()
        d_loss.backward()
        optimizer_D.step()

        z = torch.randn(batch_size, latent_dim, 1, 1).to(device)    # Training the generator
        fake_imgs = generator_forward(z)
        outputs = discriminator_forward(fake_imgs)
        g_loss = criterion(outputs, real_labels)
        optimizer_G.zero_grad()
        g_loss.backward()
        torch.nn.utils.clip_grad_norm_(generator_DCGAN.parameters(), max_norm=0.5)
        optimizer_G.step()

        G_losses.append(g_loss.item())
        D_losses.append(d_loss.item())

    print(f"Epoch [{epoch+1}/{epochs}]  Loss D: {d_loss.item():.4f}, Loss G: {g_loss.item():.4f}")
    # Visualizing output fro every 10 epochs
    if (epoch + 1) % 10 == 0:
        with torch.no_grad():
            generated_imgs = generator_forward(fixed_z).cpu()
            generated_imgs = (generated_imgs + 1) / 2  # Rescale to [0, 1]
            grid = torchvision.utils.make_grid(generated_imgs, nrow=4, padding=2, normalize=True)
            plt.figure(figsize=(8, 8))
            plt.imshow(np.transpose(grid, (1, 2, 0)))
            plt.title(f"Generated Images at Epoch {epoch+1}")
            plt.axis("off")
            plt.show()

```

```python
# Plotting the loss curves
plt.figure(figsize=(10, 5))
plt.plot(G_losses, label="Generator Loss")
plt.plot(D_losses, label="Discriminator Loss")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Comparing generated images to real CIFAR-10 images
with torch.no_grad():
    real_imgs, _ = next(iter(dataloader))
    real_imgs = (real_imgs + 1) / 2  
    grid_real = torchvision.utils.make_grid(real_imgs[:16], nrow=4, padding=2, normalize=True)
    plt.figure(figsize=(8, 8))
    plt.imshow(np.transpose(grid_real, (1, 2, 0)))
    plt.title("Real CIFAR-10 Images")
    plt.axis("off")
    plt.show()

    generated_imgs = generator_forward(fixed_z).cpu()
    generated_imgs = (generated_imgs + 1) / 2  
    grid_fake = torchvision.utils.make_grid(generated_imgs, nrow=4, padding=2, normalize=True)
    plt.figure(figsize=(8, 8))
    plt.imshow(np.transpose(grid_fake, (1, 2, 0)))
    plt.title("Generated Images")
    plt.axis("off")
    plt.show()
```
